```python
import pandas as pd
```

```python
df = pd.read_csv('./spreadsheets/reflacx_clinical.csv')
```

```python
all_genders = ["F", "M"]
all_age_gaps = [18, 30, 40, 50, 60, 70, 80, 90, 100]
num_each_group = 5
samples = []

for g in all_genders:
    for i in range(len(all_age_gaps) - 1):
        # sample
        
        t_df = df[
            (df["gender"]
            == g) & (df["age"]
            >= all_age_gaps[i]) & (df["age"]
            < all_age_gaps[i + 1])
        ].sample(n=5)
        samples.append(t_df)
```

```python
sampled_df = pd.concat(samples, axis=0)
sampled_df
```

```python
sampled_df.to_csv("./spreadsheets/gender-age-balance.csv")
```
