```python
results = []
all_res = []
all_prompt = []
all_reasons = []
for idx, data in sample_df.iterrows():
    prompt, res, result, reason = get_possible_values(
        data,
        identified_keywords,
        responses=responses,
        using_sys_p=False,
    )
    all_prompt.append(prompt)
    all_res.append(res)
    results.append(result)
    all_reasons.append(reason)
```
